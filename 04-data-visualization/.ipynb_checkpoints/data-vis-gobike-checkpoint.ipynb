## Data Gathering


```python
# make gobike data directory 
folder_name = 'gobike-data'

if not os.path.exists(folder_name):
    os.makedirs(folder_name)

for i in range(1, 13):
    """
    download and extract '/tripdata.csv.zip' files for 2018
    gobike_url = 'https://s3.amazonaws.com/fordgobike-data/201801-fordgobike-tripdata.csv.zip'

    """
    str_month = str(i).zfill(2)
    gobike_url = 'https://s3.amazonaws.com/fordgobike-data/2018' + str_month + '-fordgobike-tripdata.csv.zip'
    print(gobike_url)
    response = requests.get(gobike_url)
            
    # open zip file; return a readable and writeable view of the contents;
    zipDoc = zipfile.ZipFile(BytesIO(response.content))
        
    # puts extracted zip file into folder trip_data_files
    zipDoc.extractall(folder_name)
```

__Merge and Save Data__

```python
# merge csv files into single file and write files to disc
folder_name = 'gobike-data'

file_path = os.getcwd() + '/' + folder_name
all_files = glob.glob(file_path + '/*.csv')

temp_list = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    temp_list.append(df)
    
gobike = pd.concat(temp_list, axis=0, ignore_index=True)
```

### Ride Duration

```python
# avg ride duration
min_trip = round(np.min(sf_bikes.duration_sec), 0)
avg_trip = round(np.mean(sf_bikes.duration_sec), 0)
max_trip = round(np.max(sf_bikes.duration_sec), 0)

intervals = (
    ('weeks', 604800),  # 60 * 60 * 24 * 7
    ('days', 86400),    # 60 * 60 * 24
    ('hours', 3600),    # 60 * 60
    ('minutes', 60),
    ('seconds', 1),
    )

def display_time(seconds, granularity=2):
    result = []

    for name, count in intervals:
        value = seconds // count
        if value:
            seconds -= value * count
            if value == 1:
                name = name.rstrip('s')
            result.append("{} {}".format(value, name))
    return ', '.join(result[:granularity])
```

```python
# shortest ride
print('shortest ride: {}'.format(display_time(int(min_trip))))
```

```python
# average ride
print('average ride: {}'.format(display_time(int(avg_trip))))
```

```python
# longest ride
print('longest ride: {}'.format(display_time(int(max_trip))))
```

```python
#- ride duration histograms

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18,8))

ax1.hist(sf_bikes.duration_sec)
ax1.grid()
ax1.set_xlabel('Ride Duration (sec)');
ax1.set_ylabel('Duration Frequency')
ax1.set_title('Ride Duration Distribution')

bins = 10 ** np.arange(1, 5 + 0.1, 0.1)
ax2.hist(data=sf_bikes, x='duration_sec', bins=bins)
ax2.set_xscale('log')
ax2.grid()
ax2.set_xlabel('Duration (log10 sec)');
ax2.set_title('Ride Duration Distribution - Log Scale')

plt.subplots_adjust(wspace=0.2);

# save as png
plt.savefig('plots/ride-duration-dist.png', bbox_inches='tight')
```

__Ride Duration Comments__

The plot of ride duration data shows a strongly positively skewed distribution. Plotting the log base 10 transformation of ride duration shows a fairly normal distribution. Using a non-linear re-expression of ride duration may be useful for more sophisticated statistical analysis such as least squares.

Most rides are less than less 30 minutes long. This makes sense considering the pricing structure of GoBike, which charges subscribers in increments of 15 or 30 minutes, which encourages users to return bikes promptly and therefore increases the liklihood that bikes will be available for other riders.

This reflects the systems intended use as an alternative form of transporation and commute. It would be an interesting question to ask how many car trips this number of users represents.

### Rides Per Month


```python
# rides per month
sf_bikes['start_month'] = sf_bikes.start_time.apply(lambda x: x.month)

# group data by month
sf_bikes_monthly = sf_bikes.groupby('start_month')
```

```python
# sum and sort rides by month, reindex series
sort_month = sf_bikes.start_month.value_counts().sort_index();
sort_month.rename({1:'January', 2: 'February', 3: 'March', 
                   4: 'April', 5: 'May', 6: 'June', 7: 'July', 
                   8: 'August', 9: 'September', 10: 'October', 
                   11: 'November', 12: 'December'}, inplace=True)
sort_month.index
```

```python
# bar plot rides by month
fig, ax = plt.subplots(figsize=(14, 10))
splot = sns.barplot(y=sort_month, x=sort_month.index, palette='GnBu_d')

for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', 
                   va = 'center', xytext = (0, 10), textcoords = 'offset points')

ax.set(xlabel='Month', ylabel='ride starts');
plt.title('Monthly Ride Volume', fontsize=15)
plt.xticks(rotation = 15);
plt.ylabel('Rides Per Month');

# save as png
plt.savefig('plots/monthly-volume.png', bbox_inches='tight')
```

__Monthly Ride Volume__

This plot shows seasonal trends in system use. It's highly likely that these trends are driven by weather and temperature levels. People are more willing to commute by bicycle when the weather is good. This information can inform system planners about the number of bikes needed during different times of year and schedule system maintenence.

### Start Stations

```python
# make dataframe from counts of unique values
df_magnitude = pd.DataFrame(sf_bikes['start_station_name'].value_counts())
df_magnitude.head(3)
```

```python
# rename column: start_magnitude
df_magnitude.rename(columns={'start_station_name': 'start_magnitude'}, inplace=True)
#df_magnitude.head(3)
```

```python
# reset index, adds old index as column, to be renamed
df_magnitude.reset_index(inplace=True)
#df_magnitude.head(3)
```

```python
# rename column to names
df_magnitude.rename(columns={'index': 'start_st_name'}, inplace=True)
#df_magnitude.head(3)
```

```python
# add magnitude ratio column for colour map (unique station count / all station count)
df_magnitude['magnitude_ratio'] = df_magnitude['start_magnitude'] / len(sf_bikes['start_station_name'])
#df_magnitude.head()
```

```python
# create color map
df_colours = pd.DataFrame(cm.rainbow(np.linspace(0, 1, len(df_magnitude))));
#df_colours.head()
```

```python
# concatenate df_magnitude and df_colour
df_magnitude = pd.concat([df_magnitude, df_colours], axis=1)
#df_magnitude.head()
```

```python
# rename colour columns
df_magnitude.rename(columns={0: 'r', 1: 'g', 2: 'b', 3: 'a'}, inplace=True)
#df_magnitude.head()
```

```python
# make tuple RGBA column
df_magnitude['RGBA'] = list(zip(df_magnitude.r, df_magnitude.g, df_magnitude.b, df_magnitude.a))
df_magnitude.head()
```

```python
# sort magnitude dataframe by station name
df_magnitude.sort_values(by=['start_st_name'], inplace=True)
#df_magnitude.head()
```

```python
# reset index, adds old index as column, to be renamed
df_magnitude.reset_index(drop=True, inplace=True)
#df_magnitude.head()
```

```python
# sort geo_sf_stations by start station name
geo_sf_stations.sort_values(by=['start_station_name'], inplace=True)
#geo_sf_stations.head()
```

```python
# reset index, adds old index as column, to be renamed
geo_sf_stations.reset_index(drop=True, inplace=True)
#geo_sf_stations.head()
```

```python
# concatenate df_geo and df_mag
df_geo_station = pd.concat([geo_sf_stations, df_magnitude], axis=1)
#df_geo_station.head(3)
```

```python
# create marker size variable
df_geo_station['marker_size'] = df_geo_station.magnitude_ratio*10000
df_geo_station.head(3)
```

### Start Station Demand Plot
* Marker color and size depend on total station use
* Less use is indicated by smaller and blue markers
* Greater use is indicated by larger and red markers

```python
# plot start station markers
cmap = cm.get_cmap('cool')

fig, ax = plt.subplots(figsize = (12, 12))
sf_map.plot(ax=ax, alpha = 0.3, color='grey')
df_geo_station.plot(ax=ax, markersize='marker_size', column='magnitude_ratio', cmap='bwr', marker='o', 
                    label='GoBike Stations \n red = high demand \n blue = low demand ')
plt.title("GoBike Start Station Demand", fontsize=15)
plt.legend(loc='upper left', prop={'size':12});

# save as png
plt.savefig('plots/start_station-intensity.png', bbox_inches='tight')
```

## Conclusions

Planning for the transporation needs of the present and future is an important task for cities. Cities need effective transportation systems and cannot rely entirely on cars as much as in the past. Other forms of transportation including public bike share programs can also help alleviate some of the issues of congestion and pollution that over-dependence on cars has caused.

This analysis has described the one public bike share system in San Francisco over the course of 2018. The Ford GoBike system attracted nearly a million and a half riders during 2018, the majority of them annual subscribers to the system and the majority of them male. Plots of temporal system use shows that most riders use the system for commute to and from work in the morning and afternoon, Monday through Friday. This analysis also visually described the geographic demand for starting and ending dock stations.

Quantitative and visual descriptions of municipal transportation systems can assist city planners design and manage bike share systems more effectively.
